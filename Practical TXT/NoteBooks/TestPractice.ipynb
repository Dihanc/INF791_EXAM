```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

# Step 1: Create an instance of LabelEncoder
encoder = LabelEncoder()

# Step 2: List of columns to label encode
columns_to_encode = ['Timestamp', 'Gender', 'Status', 'Depression', 'Anxiety', 'Panic',
 'Treatment']

# Step 3: Apply LabelEncoder to each column using a loop
for col in columns_to_encode:
    df[col] = encoder.fit_transform(df[col])
```

```python
df.head()
```

```python
# Create DataFrame


# Step 1: Find the course with the highest CGPA
course_with_highest_cgpa = df[df['CGPA'] == df['CGPA'].max()]['Course'].values[0]

# Step 2: Filter the DataFrame to include only the rows with the highest CGPA course
df_filtered = df[df['Course'] == course_with_highest_cgpa]

# Step 3: Set the figure size
plt.figure(figsize=(12, 6))

# Step 4: Create a bar plot to show the distribution of CGPA per Timestamp for the course with the highest CGPA
sns.barplot(data=df_filtered, x='Timestamp', y='CGPA', palette='Set1')

# Step 5: Add labels and title
plt.xlabel('Timestamp')
plt.ylabel('CGPA')
plt.title(f'Timestamp Distribution for {course_with_highest_cgpa} with Highest CGPA')

# Show the plot
plt.grid(axis='y')
plt.show()
```

```python
# Group by 'Course' and sum the 'Depression' values
depression_sum = df.groupby('Course')['Depression'].sum()

# Find the course with the most depressed students
most_depressed_course = depression_sum.idxmax()
print(f"The course with the most depressed students is '{most_depressed_course}'")
```

```python
# Filter the DataFrame for the course with the most depressed students
df_filtered = df[df['Course'] == most_depressed_course]

# Group by 'Gender' within the filtered data and sum the 'Depression' values
gender_depression_sum = df_filtered.groupby('Gender')['Depression'].sum()

# Find the gender with the highest depression sum
most_depressed_gender = gender_depression_sum.idxmax()
print(f"The most depressed gender among students in the '{most_depressed_course}' course is '{most_depressed_gender}'")
gender_mapping = {0: 'Female', 1: 'Male'}
most_depressed_gender = gender_mapping[most_depressed_gender]
print(f"The most depressed gender among students in the '{most_depressed_course}' course is '{most_depressed_gender}'")
```

```python
students_wanting_treatment = df[df['Treatment']== 1]
```

```python
# Step 2: Group the filtered data by 'Course' and 'Gender' and calculate the count of students wanting treatment
treatment_counts = students_wanting_treatment.groupby(['Course', 'Gender']).size().reset_index(name='Count')
```

```python
# Step 3: Find the course and gender with the most students wanting treatment
most_wanted_course_gender = treatment_counts[treatment_counts['Count'] == treatment_counts['Count'].max()]
```

```python
# Step 4: Extract the course and gender with the most students wanting treatment
most_wanted_course = most_wanted_course_gender['Course'].values[0]
most_wanted_gender = most_wanted_course_gender['Gender'].values[0]
```

```python
# Step 5: Print the result
print(f"The course and gender that need the most treatment are '{most_wanted_course}' and '{most_wanted_gender}'.")
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

# Generate random data
data = np.random.normal(loc=3.5, scale=0.5, size=1000)

# Plot the histogram
plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, density=True, alpha=0.6, color='skyblue')

# Fit a normal distribution to the data
mean, std_dev = norm.fit(data)

# Plot the normal distribution curve
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
y = norm.pdf(x, mean, std_dev)
plt.plot(x, y, 'r--', linewidth=2)

# Add labels and show the plot
plt.title('Histogram with Normal Distribution Fit')
plt.xlabel('Data Values')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X = df.drop('MEDV', axis = 1)
Y = df['MEDV']
X_train, X_test, y_train, y_test = train_test_split(X, Y,
train_size=0.8, random_state=42, shuffle = True)
```

```python
from sklearn.linear_model import LinearRegression

# Use Linear Regression for continuous target values
model = LinearRegression()
model.fit(X_train, y_train)
regression_preds = model.predict(X_test)
```

```python
# Display the intercept and coefficients
print("Intercept:", model.intercept_)
print("Coefficients:", model.coef_)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score


# Step 6.1: Make predictions
train_preds = model.predict(X_train)
test_preds = model.predict(X_test)

# Step 6.2: Calculate R score (a measure of predictive accuracy)
train_r2 = r2_score(y_train, train_preds)
test_r2 = r2_score(y_test, test_preds)

# Display predictive accuracy
print(f"Training R Score: {train_r2:.2f}")
print(f"Testing R Score: {test_r2:.2f}")

# Calculate Mean Squared Error for additional evaluation
train_mse = mean_squared_error(y_train, train_preds)
test_mse = mean_squared_error(y_test, test_preds)

print(f"Training Mean Squared Error: {train_mse:.2f}")
print(f"Testing Mean Squared Error: {test_mse:.2f}")
```
